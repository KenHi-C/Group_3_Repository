```python
# Solution - enter your code solution below
from datetime import date, timedelta
import calendar
import random

#Caregiver class
class Caregiver:
    #Initializes caregiver attributes
    def __init__(self, name, phone=None, email=None, pay_rate=0.0, hours=0, availability=None):
        self.name = name
        self.phone = phone
        self.email = email
        self.pay_rate = pay_rate
        self.hours = hours
        self.availability = availability
        
    #Adds hours to caregiver
    def add_hours(self, hours):
        self.hours += 6
    
    #Returns number of shifts worked per caregiver
    def get_num_shifts(self):
        if self.hours > 0:
            return self.hours/6
        else:
            return 0
    
    #Returns caregiver availibility for a certain shift
    def get_availability(self, date, shift):
        return self.availability.get(date, {}).get(shift, 'unavailable')
    
    def __repr__(self):
        return f"Caregiver({self.name}, {self.phone}, {self.email}, Pay Rate: ${self.pay_rate:.2f}, Hours: {self.hours})"

#Pay
class Pay(Caregiver, Schedule):
    def __init__(self, caregiver):
        self.caregiver = caregiver
        
    #Define a class for calculating the monthly pay.
    def calculate_monthly_pay(self):
        weekly_pay = self.caregiver.hours * self.caregiver.pay_rate
        monthly_pay = 4 * weekly_pay
        monthly_shifts = 4 * self.caregiver.get_num_shifts()
        weekly_report = f"Monthly Pay Report:\n{self.caregiver.name}: {monthly_shifts} shifts, ${monthly_pay}\n"
        return weekly_report
    
    #Define a class for calculating the weekly pay.
    def calculate_weekly_pay(self):
        weekly_pay = self.caregiver.hours * self.caregiver.pay_rate
        weekly_report = f"Weekly Pay Report:\n{self.caregiver.name}: {self.caregiver.get_num_shifts()} shifts, ${weekly_pay}\n"
        return weekly_report

#Schedule Class 
class Schedule:
    #Initializes schedule with startdate, schedule dictionary, list of caregivers, and shifts
    def __init__(self, start_date):
        self.start_date = start_date
        self.schedule = dict()
        self.caregivers = []
        self.shifts = ['AM', 'PM']
    
    #Adds a new caregiver to the caregivers class
    def add_caregiver(self, caregiver):
        if len(self.caregivers) <= 8:   #Limits caregiver list to 8 caregivers
            self.caregivers.append(caregiver)
        else:
            print("There are already 8 caregivers working.")
    
    #Generates caregiver's schedules based on availability and available shifts
    #Asks for start date and provides the schedule for the next week
    def generate_schedule(self):
        current_date = self.start_date
        random.shuffle(self.caregivers)  #Randomizes order to prioritize different caregivers each time
        for _ in range(7):  #Generates for every day in the week
            for shift in self.shifts:
                caregiver = self.assign_shift(current_date, shift) #References assign_shift() method to find best fit
                if caregiver:
                    self.schedule[(current_date, shift)] = caregiver.name
                    caregiver.add_hours(6)  
            current_date += timedelta(days=1)  #Increments the date by one day
        
    #Method to assign the ideal shifts for caregivers
    def assign_shift(self, current_date, shift):
        date_str = current_date.strftime('%Y-%m-%d')  # Convert date to string
        for caregiver in self.caregivers:
            availability = caregiver.get_availability(date_str, shift)
            if availability == 'preferred':
                return caregiver
            elif availability == 'available':
                return caregiver
        return None

    def display_schedule(self):
        # Start and end date of the week
        end_date = self.start_date + timedelta(days=6)
        
        # Header for the HTML
        html_schedule = f"""
        <html>
        <head>
            <title>Work Schedule: {start_date} to {end_date}</title>
            <style>
                table {{
                    border-collapse: collapse;
                    width: 80%;
                    margin: 20px auto;
                    font-family: Arial, sans-serif;
                }}
                th, td {{
                    border: 1px solid #ddd;
                    padding: 8px;
                    text-align: center;
                }}
                th {{
                    background-color: #f4f4f4;
                }}
                tr:nth-child(even) {{ background-color: #f9f9f9; }}
                h1 {{
                    text-align: center;
                }}
            </style>
        </head>
        <body>
            <h1>Work Schedule: {self.start_date.strftime('%b %d, %Y')} to {end_date.strftime('%b %d, %Y')}</h1>
            <table>
                <tr>
                    <th>Date</th>
                    <th>Day</th>
                    <th>AM Shift</th>
                    <th>PM Shift</th>
                </tr>
        """
        
        # Generate rows for each day in the week
        current_date = self.start_date
        while current_date <= end_date:
            day_name = calendar.day_name[current_date.weekday()]
            am_caregiver = self.schedule.get((current_date, 'AM'), "Unassigned")
            pm_caregiver = self.schedule.get((current_date, 'PM'), "Unassigned")
            
            html_schedule += f"""
                <tr>
                    <td>{current_date.strftime('%Y-%m-%d')}</td>
                    <td>{day_name}</td>
                    <td>{am_caregiver}</td>
                    <td>{pm_caregiver}</td>
                </tr>
            """
            current_date += timedelta(days=1)
        
        html_schedule += """
            </table>
        </body>
        </html>
        """
        
        # Write to an HTML file
        file_path = "weekly_schedule.html"
        with open(file_path, "w") as file:
            file.write(html_schedule)

        # Automatically open in the default web browser
        abs_file_path = os.path.abspath(file_path)
        webbrowser.open(f"file://{abs_file_path}")
        
        return html_schedule
```
